```python
import itertools
def create_pattern_for_subsystem(num_components,level_num):
    all_possible_values = [range(level_num) for _ in range(num_components)]
    print(all_possible_values)
    all_cases = list(itertools.product(*all_possible_values))
    
    return all_cases
# Example with N=3 and P=2
N = 5
P = 5

all_cases = create_pattern_for_subsystem(N, P)
all_cases2= create_pattern_for_subsystem(4, P)
all_cases3 = create_pattern_for_subsystem(3,P)
# Print all cases
for case in all_cases3:
    print(case)
```

```python
maintenance_level={0:{"cost":0,"time":0,"rul_increase":0},1:{"cost":30,"time":0.8,"rul_increase":0.2},2:{"cost":75,"time":2,"rul_increase":0.4},3:{"cost":150,"time":3.5,"rul_increase":0.8},4:{"cost":250,"time":5,"rul_increase":1}}

compenent_RUL={ 0:0.893,1:0.003,2:0.000,3:0.003,4:0.000}
compenent_RUL2={ 0:0.860,1:0.000,2:0.000,3:0.00}
compenent_RUL3={ 0:0.770,1:0.003,2:0.000}
def calcul_cost_time_all_case(all_cases,maintenance_level,compenent_RUL):
    cost_time_case={}
    for case in all_cases:
        cost=0
        time=0
        rul_after_maintenance_list={}
        reliability_subsystem=1
        num_compenent=0
        for compenent  in case :
    
            if(compenent!='(' and compenent!=',' and compenent!=')' and compenent!=' '):
                compenent=int(compenent)
                compenent_level=compenent
                time=time+maintenance_level[compenent_level]["time"]
                cost=cost+maintenance_level[compenent_level]["cost"]
                rul_after_maintenance=compenent_RUL[num_compenent]+maintenance_level[compenent_level]["rul_increase"]
                reliability_subsystem=rul_after_maintenance*reliability_subsystem
                if(rul_after_maintenance>1):
                    rul_after_maintenance=1
                rul_after_maintenance_list[num_compenent]=rul_after_maintenance 
                num_compenent=num_compenent+1

        key = str(case)
        if key not in cost_time_case:
            cost_time_case[key] = {}
        cost_time_case[key]["cost"] = cost
        cost_time_case[key]["time"] = time  
        cost_time_case[key]["after_maintenance"] = rul_after_maintenance_list  
        cost_time_case[key]["reliability_subsytem"]=reliability_subsystem     
    
    return cost_time_case
all_cases=calcul_cost_time_all_case(all_cases,maintenance_level,compenent_RUL) 
all_cases2=calcul_cost_time_all_case(all_cases2,maintenance_level,compenent_RUL2) 
all_cases3=calcul_cost_time_all_case(all_cases3,maintenance_level,compenent_RUL3) 
all_cases3
```
